```python
%matplotlib inline
import pymc
import seaborn
import numpy as np
import scipy
import math
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import itertools
import matplotlib.gridspec as gridspec
```

```python
# illustration of propgation of uncertainty of variables in non-linear transformatio
beta = pymc.Exponential('beta', 0.5, 4)
alpha = pymc.Normal("alpha", 0.1, 3)
@pymc.deterministic
def transform(beta=beta, alpha=alpha):
    alpha = np.expand_dims(alpha, axis=0)
    beta = np.expand_dims(beta, axis=0)
    return math.log(abs(alpha*beta))
model = pymc.Model([transform, alpha, beta])
sampler = pymc.MCMC(model)
samples = sampler.sample(iter=1000)
seaborn.jointplot(sampler.trace('beta')[100:], sampler.trace('alpha')[100:], kind='kde', color='g', ratio=3)
```

```python
seaborn.distplot(sampler.trace('transform')[100:], color='g')
```

```python
# illustrate all 64 possible combinations of multiplicity terms for one torsion
array = []

# use itertools to get all possible combinations
for i in itertools.product((0,1), repeat=6):
    array.append(i)
    
grid = np.ndarray((8,8), dtype='6int8')
n = 0
for i in range(8):
    for j in range(8):
        grid[i][j] = array[n]
        n +=1
        
import matplotlib.gridspec as gridspec

new_style = {'grid': False}
plt.rc('axes', **new_style)
fig, axes = plt.subplots(nrows=1, ncols=8, figsize=[10,10])
# To get rid of white space between subplots
gs1 = gridspec.GridSpec(1, 8)
gs1.update(wspace=0, hspace=0) #


n = 0
for i in range (8):
    axes[ i] = plt.subplot(gs1[i])       
    axes[i].imshow(grid[n], interpolation='nearest', extent=[0, 6, 0, 8])
    axes[i].grid(True, linestyle='-', color='y', which='major')
    for m in range(9):
        axes[i].axhline(m, color='blue', lw=2)
    axes[i].axvline((0), color='red', lw=3)
    axes[i].axvline(6, color='red', lw=3)
    axes[i].axes.xaxis.set_ticklabels([])
    axes[i].axes.yaxis.set_ticklabels([])
        
    n +=1

```
